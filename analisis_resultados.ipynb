```python
import pandas as pd
```

Lectura de los resultados

```python
results = pd.read_csv('results1.csv')
results = results.drop(columns=['Unnamed: 0'])
ulutas_results = pd.read_csv('ulutas_results.csv')
```

Agregamos resultados

```python
agg_results = results.groupby(["Problem"]).mean().reset_index() #agrego por problema y calculo media de eficacia, iteraciones y tiempo de ejecuci√≥n
agg_results = agg_results.join(ulutas_results.set_index('Problem'), on = 'Problem') #hago join con fichero de resultados de ulutas
agg_results = agg_results[['Problem', 'Problem Code', 'Efficacy', 'Iterations', 'Elapsed Time', 'Ulutas Efficacy']] #me quedo con columnas que me interesan
agg_results['Efficacy Difference'] = round((agg_results['Ulutas Efficacy'] - agg_results['Efficacy'])*100/agg_results['Ulutas Efficacy'], 2) #calculo diferencia entre nuestra eficacia y la de Ulutas
agg_results
```
